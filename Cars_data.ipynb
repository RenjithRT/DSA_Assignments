### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
# Assign gender wise calculation to the variables
male_count=cars_data[['Buyer Gender']].value_counts()['Male']
female_count=cars_data[['Buyer Gender']].value_counts()['Female']
print('Total male and female buyers participated in the sales count: ',male_count,'and',female_count)
```

```python
plt.figure(figsize=(12,6))
plot=plt.subplot(1,1,1)

male_count=cars_data[['Buyer Gender']].value_counts()['Male']
female_count=cars_data[['Buyer Gender']].value_counts()['Female']

plt.bar(cars_data['Buyer Gender']=='Male', male_count, width=0.4, align='center', label='Male')
plt.bar(cars_data['Buyer Gender']=='Female', female_count,width=0.3, align='edge', label='Female')

#get handles and labels

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(labels)

plt.title('Male vs Female buyers participated in the sales', fontweight='bold')
plt.xticks([])
plt.grid(linestyle='--')
plt.xlabel('Gender')
plt.ylabel('Sales units in number')
plt.show()
```

From the above bar plot, we can observe female buyers participated in the sales count has more than the male buyers participated in the sales count. Total male and female buyers participated in the sales count are 4948 and 5052.

### 4. Find the top 5 cars based on their sales price.

First, we have to sort the values based on their sales price and then display the top 5 headings using head() method.

```python
# This is equvalent to cars_data['Sale Price'].nlargest(5).index
top_5_cars=cars_data.sort_values('Sale Price', ascending=False).head() # here ascending is assigned false so it perform decending order
top_5_cars
```

```python
top_5_cars = top_5_cars[['Make','Model','Sale Price']] # Display the car brands with sale price
top_5_cars.head()
```

The top five cars are Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES. The Volkswagen Jetta III is the top one out of five with sale price 99997.65 and Lexus where model ES is the top of 5th position with sale price 99973.44.

### 5. Find the least 5 cars based on their Resell price.

In the above session we displayed the top 5 cars using head() method , similarly we can diplay the least 5 cars with  ascending as true.

```python
# This is equvalent to cars_data['Resell Price'].nsmallest(5).index
least_5_cars=cars_data.sort_values('Resell Price', ascending=True).head() 
least_5_cars
```

```python
least_5_cars = least_5_cars[['Make','Model','Resell Price']] # Display the car brands with resale price
least_5_cars.head()
```

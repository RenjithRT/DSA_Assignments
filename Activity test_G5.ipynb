```python
# checking the top 10 population country in the dataset
top_10 = df.sort_values('Population in thousands (2017)', ascending=False).head(10)
top_10
```

```python
# Data visualization of the top 10 population country in the dataset

fig=plt.subplots(figsize=(16,4))
x = range(10)
plt.bar(x, top_10['Population in thousands (2017)'] / 10 ** 4)

plt.xticks(x, top_10.country, rotation = 'vertical') #plt.xticks(coordinates, actual string values, show this lables vertically)

plt.title('10 Most Populous Countries in 2017')
plt.ylabel('Population in thousands')
plt.show()
```

The above chart shows that, China is having the highest with above 140 thousand populations and India is the second highest with above 130 thousand populations.

```python
# checking the least 10 population country in the dataset
least_10 = df.sort_values('Population in thousands (2017)', ascending=True).head(10)
least_10
```

```python
# Data visualization of the least 10 population country in the dataset

fig=plt.subplots(figsize=(16,4))
x = range(10)
plt.bar(x, least_10['Population in thousands (2017)'])

plt.xticks(x, least_10.country, rotation = 'vertical') #plt.xticks(coordinates, actual string values, show this lables vertically)

plt.title('10 Least Populous Countries in 2017')
plt.ylabel('Population in thousands')
plt.show()
```

This chart shows that, Montserrat is having the lowest with above 5 thousand populations and Tuvalu is the second lowest with above 11 thousand populations.

```python
# Data visualization of the Region wise Population in the dataset

fig=plt.subplots(figsize=(16,4))
plt.xticks(rotation=90)
sns.barplot(x='Region',y=df['Population in thousands (2017)'] / 10**4,data=df)
plt.title('Region wise Population')
plt.show()
```

The region wise population, where Eastern Asia is the highest one and second is the Southern Asia. The lowest population in regions are Caribbean, Polynesia, Melanesia and Micronesia.

```python
# Data visualization of the Population in Surface area in the dataset

plt.figure(figsize=(14,6))
plt.scatter(df['Surface area (km2)'],df['Population in thousands (2017)']/10**4,s=14,c='green',marker='^')
plt.title('Population in Surface area')
plt.xlabel('Surface area (km2)')
plt.ylabel('Population in thousands (2017)')
plt.show()
```

As you can see that less surface area with high populations and high surface area with less populations

### 4. On average which region has the highest

**a. GDP per capita (current US$)**

```python
# display the dataset
df.head()
```

```python
# calculate the mean of GDP per capita with respect to region
region_GDP = df.groupby('Region')['GDP per capita (current US$)'].mean().round(2).sort_values()
region_GDP
```

```python
#display the bottom values in the dataset
region_GDP.tail(1)
```

```python
# Data visualization of the Average of GDP per capita in the dataset

fig= plt.subplots(figsize=(16,4))
plt.xticks(rotation=90)
ax=sns.lineplot(data=region_GDP, markers=True,color='green')
ax.set(ylabel='GDP per capita Mean')
ax.set(title='Average of GDP per capita')
plt.show()
```

From the graph, on average Northern America region has the highest GDP per capita with mean 57659

**b. International trade: Imports (million US$)**

```python
# calculate the mean of GDP per capita with respect to region
region_trade = df.groupby('Region')['International trade: Imports (million US$)'].mean().round(2).sort_values()
region_trade
```

```python
#display the bottom values in the dataset
region_trade.tail(1)
```

```python
# Data visualization of the Average of International Trade: Imports in the dataset

fig= plt.subplots(figsize=(16,4))
plt.xticks(rotation=90)
ax=sns.lineplot(data=region_trade, markers=True, color='violet')
ax.set(ylabel='International Trade: Imports Mean')
ax.set(title='Average of International Trade: Imports')
plt.show()
```

on average Northern America region has the highest International trade: imports with mean 663552.25

### 5. Find out which region has more consistent with respect to GDP per capita (current US$

```python
# calculate the standard deviation of GDP per capita with respect to region
consistent_region = df.groupby('Region')['GDP per capita (current US$)'].std().round(2).sort_values()
consistent_region
```

```python
#display the topmost two values in the dataset
consistent_region.head(2)
```

```python
# Data visualization of the Standard Deviation of GDP per capita in the dataset

fig= plt.subplots(figsize=(16,4))
plt.xticks(rotation=90)
ax=sns.lineplot(data=consistent_region, markers=True,color='blue')
ax.set(ylabel='Standard Deviation of GDP per capita' )
ax.set(title='Standard Deviation of GDP per capita')
plt.show()
```

Western Africa has more consistent with respect to GDP per capita (current US$) and Northern Africa follows the slighly same it's because closer to the Western Africa. And Eastern asia is the least consistency
